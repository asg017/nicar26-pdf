## Data Table extraction

> *Adapted from the original [pdfplumber example notebook!](https://github.com/jsvine/pdfplumber/blob/stable/examples/notebooks/extract-table-ca-warn-report.ipynb)*

Take a look at [This CA WARN Report](samples/ca-warn-report.pdf), which is a periodic report from the California Employment Development Department that notfies when a company in CA performs large layoffs. 

In the PDF, can you select the text? Yes you can!

Let's try extracting text!

````python
pdf = pdfplumber.open("samples/ca-warn-report.pdf")
print(pdf.pages[0].extract_text())
````

oh no PDFs suck :(

Fear not! pdfplumber has some utilities for extract data from tables like this!

Instead of `.extract_text()`, try `.extract_table()`:

````python
pdf.pages[0].extract_tables()
````

Oh much better! 

The data isn't 100% perfect, so you'll have to do some cleaning yourself. Here's an example:

````python
table = pdf.pages[0].extract_tables()[0]
headers = table[0]
rows = []
for row in table[1:]:
    record = dict(zip(headers, row))
    for column in ["Effective", "Received"]:
        record[column] = record[column].replace(" ", "")
    rows.append(record)
print(json.dumps(rows, indent=2))
````


pdfplumber also has a builtin table debugger for visualizing its process better:

````python
im = pdf.pages[0].to_image()
im.debug_tablefinder()
````
